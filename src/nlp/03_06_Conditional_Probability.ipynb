# 3.6 조건부 확률(Conditional Probability)



A = 학생이 남학생인 사건  
B = 학생이 여학생인 사건  
C = 학생이 중학생인 사건  
D = 학생이 고등학생인 사건  
이라고 했을 때, 아래의 확률을 계산해보겠습니다.


1. 학생을 뽑았을 때, 남학생일 확률  
P(A)=180/360=0.5

2. 학생을 뽑았을 때, 고등학생이면서 남학생일 확률  
P(A∩D) = 80/360

3. 고등학생 중 한명을 뽑았을 때, 남학생일 확률  
P(A|D) = 80/200 = P(A∩D)/P(D) = (80/360)/(200/360) = 80/200 = 0.4
