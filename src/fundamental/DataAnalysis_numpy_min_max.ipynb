```python
import numpy as np
```

```python
x = np.array([5,4,3,2,1,0])
```

```python
# min,max
print('np.min :{} , np.max : {}'.format(np.min(x),np.max(x)))
```

```python
# argmin, argmax
print('np.argmin :{} , np.argmax : {}'.format(np.argmin(x),np.argmax(x)))
```

```python
# where
np.where(x>=3)
```

```python
x[np.where(x>=3)]
```

```python
# change value
np.where(x>=3,3,x)
```
